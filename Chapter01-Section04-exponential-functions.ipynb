```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

- Example of a reflection of f(x) = 1/x and f(x) = 1/x^2 about the x and y axes. Remember you have to manage the domain. 

```python
a = 1
npoints = 99
xpos = np.linspace(0.01, a, npoints)
xneg = -xpos
```

```python
n = 1

y0 = (1/xpos)**(n)
y1 = (-1/xpos)**(n)
y2 = (1/xneg)**(n)
y3 = (-1/xneg)**(n) 
```

- Add a title, labels, grid, labels, and legend

```python
plt.plot(xpos, y0, label=r'$(1/xpos)^n, [0.1, 1.0]$', color='black')
plt.plot(xpos, y1, label=r'$(-1/xpos)^n, [0.1, 1.0]$', color='red')
plt.plot(xneg, y2, label=r'$(1/xneg)^n [-1.0, -0.1]$', color='green')
plt.plot(xneg, y3, label=r'$-(1/xneg)^n [-1.0, -0.1]$', color='blue')
plt.title('X,Y reflections')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
n = 2

y0 = (1/xpos)**(n)
y1 = (-1/xpos)**(n)
y2 = (1/xneg)**(n)
y3 = (-1/xneg)**(n) 
plt.plot(xpos, y0, label=r'$(1/xpos)^n, [0.1, 1.0]$', color='black')
plt.plot(xpos, y1, label=r'$(-1/xpos)^n, [0.1, 1.0]$', color='red')
plt.plot(xneg, y2, label=r'$(1/xneg)^n [-1.0, -0.1]$', color='green')
plt.plot(xneg, y3, label=r'$-(1/xneg)^n [-1.0, -0.1]$', color='blue')
plt.title('X,Y reflections')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Template for plotting multiple functions on a common graph

# Define your x range

x = np.linspace(-2,2,  400)  # Adjust range and number of points as needed

# Define your functions here
y1 = 2**x          # Example: y = 2^x
y2 = 10**x         # Example: y = 10^x
y3 = (1/3) ** x    # Example: y = (1/3)^x
y4 = (1/10) ** x   # Example: y = (1/10)^x
plt.plot(x, y1, label='y = 2^x')
plt.plot(x, y2, label='y = 10^x')
plt.plot(x, y3, label='y = (1/3)^x')
plt.plot(x, y4, label='y = (1/10)^x')

plt.title('Section 1.4 #7')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_neg_exp = np.exp(-xpos)
plt.plot(xpos, y_neg_exp, label='y = e^{-x}', color='orange')
plt.title('y = e^{-x}')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
