```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.pi/2
npoints = 100
x = np.linspace(0, np.pi, npoints)
y = np.cos(x)
x_inv = np.linspace(-1.0, 1.0, npoints)
y_inv = np.arccos(x_inv)
```

```python
plt.plot(x, y, label='f(x) = sin(x)', color='red')
plt.plot(x_inv, y_inv, label='f(x) = arcsin(x)', color='green')
plt.plot(x, x, label="line of symmetry", color='grey', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')
plt.title('cos(x) and arccos(x)')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
