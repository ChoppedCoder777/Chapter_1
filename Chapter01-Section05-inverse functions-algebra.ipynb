```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x_vals = np.linspace(0, 1, 400)
y_vals = 1 - x_vals**2

y_vals_inv = np.linspace(0, 1, 400)
x_vals_inv = np.sqrt(1 - y_vals_inv)

plt.plot(x_vals, y_vals, label='f(x) = 1 - x²', color='red')
plt.plot(y_vals_inv, x_vals_inv, label='Inverse: f⁻¹(y)', color='green')
plt.plot(x_vals, x_vals, color='black', linestyle=':', label='y = x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and Its Inverse')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
sym.init_printing()
x, y = sym.symbols('x y')
```

```python
f = (1+ sym.exp(-x))
eq = sym.Eq(y, f)
inverse_solved = sym.solve(eq, x)
print ("equation")
display (eq)
print("inverse")
display (inverse_solved)
```

```python
inverse_solved = sym.solve(eq, x)[0]
```

```python
f_lambdified = sym.lambdify(x, f, 'numpy')
inverse_lambdified = sym.lambdify(y, inverse_solved, 'numpy')
```

```python
x_vals = np.linspace(0.01, 10, 400)
y_vals = f_lambdified(x_vals)
inverse_vals = inverse_lambdified(y_vals)
```

```python
plt.plot(x_vals, y_vals, label='f$(x)$', color='red')
plt.plot(y_vals, inverse_vals, label='f$^{-1}(x)$', color='blue')
sym_min = min(np.min(x_vals), np.min(y_vals), np.min(inverse_vals))
sym_max = max(np.max(x_vals), np.max(y_vals), np.max(inverse_vals))
plt.plot([sym_min, sym_max], [sym_min, sym_max], color='black', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and Its Inverse')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
